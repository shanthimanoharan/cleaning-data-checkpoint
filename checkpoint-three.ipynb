# Checkpoint Three: Cleaning Data

Now you are ready to clean your data. Before starting coding, provide the link to your dataset below.

My dataset: https://www.kaggle.com/datasets/tomaslui/healthcare-dataset

Import the necessary libraries and create your dataframe(s).

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Create the dataframes for all the 9 Files
# DimCptCode
# DimDate
# DimDiagnosisCode
# DimLocation
# DimPatient
# DimPhysician
# DimPayer
# DimTransaction
# FactTable

#read in dataset
df_fact = pd.read_csv('dataset/csv/FactTable.csv')
df_location = pd.read_csv('dataset/csv/DimLocation.csv')
df_patient = pd.read_csv('dataset/csv/DimPatient.csv')
df_physician = pd.read_csv('dataset/csv/DimPhysician.csv')
df_date = pd.read_csv('dataset/csv/DimDate.csv')
df_cptcode = pd.read_csv('dataset/csv/DimCptCode.csv')
df_diagnosiscode = pd.read_csv('dataset/csv/DimDiagnosisCode.csv')
df_payer = pd.read_csv('dataset/csv/DimPayer.csv')

#Merge all the dataframes
df = pd.merge(pd.merge(pd.merge(pd.merge(pd.merge(pd.merge(pd.merge(df_fact, df_location, on='dimLocationPK'), df_patient, on='dimPatientPK', suffixes=('', '_drop')), df_physician, on='dimPhysicianPK'), df_date, on='dimDateServicePK'),df_cptcode,
on ='dimCPTCodePK'),df_diagnosiscode,on='dimDiagnosisCodePK'),df_payer, on='dimPayerPK')             
#df.drop(['PatientNumber_drop'], axis=1, inplace=True)
df.info()
df.head()
```

## Missing Data

Test your dataset for missing data and handle it as needed. Make notes in the form of code comments as to your thought process.

```python
# Try to find some missing data with isna() function!
df.isna().sum()


# No missing data in the dataset
```

```python
# Finding missing data percentage which is more easy to read!
for col in df.columns:
    pct_missing = np.mean(df[col].isnull())
    print('{} - {}%'.format(col, round(pct_missing*100)))
# There is no missing data
```

## Irregular Data

Detect outliers in your dataset and handle them as needed. Use code comments to make notes about your thought process.

```python
# Keep an eye out for outliers!
print(df.drop(['FactTablePK', 'dimPatientPK', 'dimPhysicianPK', 'dimDatePostPK', 'dimDateServicePK', 'dimCPTCodePK', 'dimPayerPK', 'dimTransactionPK', 'dimLocationPK','PatientNumber', 'ProviderNpi', 'dimDiagnosisCodePK','Year','MonthPeriod','Day'], axis=1).describe())

#Found Outliers in a dataset with negative values which can indicate unusual or unexpected data points that deviate significantly from the majority of the data.
#GrossCharge and CPTUnits cannot be negative so i want to delete the rows with negative values.
df = df[df['GrossCharge'] >= 0]
df = df[df['CPTUnits'] >= 0]
df.info()
df.head()
```

## Unnecessary Data

Look for the different types of unnecessary data in your dataset and address it as needed. Make sure to use code comments to illustrate your thought process.

```python
# The column name 'PatientNumber_drop' in the DimFact Table is duplicate of the 'PatientNumber' column in the DimPatient table, so i want to drop the column.
df.drop(columns='PatientNumber_drop',inplace=True)
df.info()
df.head()

```

## Inconsistent Data

Check for inconsistent data and address any that arises. As always, use code comments to illustrate your thought process.

```python
# Look out for inconsistent data!
df['CptCode'].unique()
df['GrossCharge'].unique()
df['CPTUnits'].unique()

# I dont see any inconsistant data in the dataset.Hence no cleanup is needed.

# Export the cleaned data set to CSV
df.to_csv('dataset/csv/HealthcareDatasetClean.csv', index=False)
```
